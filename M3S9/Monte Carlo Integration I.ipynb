# Hit and Miss
Given $h(x)$ is bounded ($\forall x \in [a,b], 0 \leq g(x) \leq c$). We want to estimate 
$$\theta = \int_a^b g(x) dx$$ 
We sample $n$ points from the region $[a,b] \times [0,c]$ and see how many points fall into the region of integration.

Here is an outline for sampling procedure:
- Sample x-coordinate $u_i$ from $U_i \sim U(a,b)$.
- Calculate $g(u_i)$.
- Sample y-coordinate $v_i$ from $V_i \sim U(0,c)$.
- Compare $g(u_i)$ and $v_i$. If $g(u_i) \geq v_i$ then $(u_i, v_i)$ is in the region, otherwise $(u_i, v_i)$ is not in the region.

Thus we may estimate the area as (area of $[a,b] \times [0,c]$) $\times$ (proportion of points following into region), or  
$$\hat{\theta} = c(b-a) \frac{\sum_{i=1}^n I(g(u_i) \geq v_i)}{n}$$

## Example
We aim to find the area of the upper right quarter of the ellipse $y = 2\sqrt{1 - x^2}$ ($0 \leq x \leq 1$, $0 \leq y \leq 2$). <br>

The area is 
$$S = \int_0 ^ 1 2\sqrt{1 - x^2} dx$$

We sample using 1000 points.

```python
%pylab
```

```python
u = np.random.rand(1000) #x-coordinates
v = 2*np.random.rand(1000) #y-coordinates

# Comparison
g = 2*np.sqrt(1-u**2)
inregion = u[g >= v]

# Finding Area
S = 2*len(inregion)/1000
disp(S)
```

```python
#Plot g(x)
x = np.linspace(0,1,num=500, endpoint = False)
y = 2*np.sqrt(1-x**2)
plt.plot(x,y)
plt.scatter(u,v)
```

# Crude Monte Carlo
Consider the integral $$\theta = \int_a ^b g(x) dx$$
We may modify the integral to make it include a probability density function of distribution $X$, i.e. $f_X (x)$:
$$\theta = \int_a ^ b \frac{g(x)}{f_X (x)} f_X(x) dx = \int_a ^b \phi(x) f_X(x) dx = E_{f_X}(\phi(X))$$
Thus we may make observations from $X$, i.e. $X_1, X_2, ... X_n$ (iid with $X$), and estimate $\theta$ using the estimator
$$\hat{\theta} = \frac{1}{n} \sum_{i=1}^n \phi(X_i)$$
It is very easy to see that the estimator is unbiased. The variance of estimator is 
$$\text{Var}_{f_X}(\hat{\theta}) = \frac{1}{n} \text{Var}_{f_X}(\phi(X)) 
= \frac{1}{n} \int_a ^b (\phi(x) - \theta)^2 f(x) dx = \propto \frac{1}{n}$$
It can also be written as
$$\text{Var}_{f_X}(\hat{\theta}) = \frac{1}{n} \text{Var}_{f_X}(\phi(X)) 
= \frac{1}{n} (E((\phi(x))^2) - (E(\phi(x)))^2) = \frac{1}{n} \bigg( \int_a ^b (\phi(x))^2 f_X (x) dx - \theta^2 \bigg)$$
We wish to minimize variance by suitable choice of distribution $X$.

## Example

From the last example, the simplest way to execute Crude Monte Carlo is to let $\phi(x) = 2\sqrt{1-x^2}$ and $f_X(x) = 1$ (i.e. $X \sim U(0,1)$.)

```python
xi = np.random.rand(1000) #x-coordinates
phixi = 2*np.sqrt(1-xi**2)
S = np.sum(phixi)/1000
disp(S)
```
